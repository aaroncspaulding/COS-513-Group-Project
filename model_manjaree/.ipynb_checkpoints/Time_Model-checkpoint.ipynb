```python
import pandas as pd
import numpy as np
import torch
import pyro
import pyro.distributions as dist
from pyro.infer import SVI, Trace_ELBO
from pyro.optim import Adam
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import numpy as np
import torch
import pandas as pd
import matplotlib.pyplot as plt

# Similar to what Aaron did but didn't use Hierachial set up here
```

```python
# Column Mapping & Data Loading

census_svi_column_mapping = {
    'B01001_001E': 'Total Population',
    'B19001_001E': 'Household Income Distribution',
    'B19013_001E': 'Median Household Income',
    'B19001_002E': 'Income Bracket 1',
    'B19001_003E': 'Income Bracket 2',
    'B19001_004E': 'Income Bracket 3',
    'B19001_005E': 'Income Bracket 4',
    'B19001_006E': 'Income Bracket 5',
    'B19001_007E': 'Income Bracket 6',
    'B19001_008E': 'Income Bracket 7',
    'B19001_009E': 'Income Bracket 8',
    'B19001_010E': 'Income Bracket 9',
    'B19001_011E': 'Income Bracket 10',
    'B19001_012E': 'Income Bracket 11',
    'B19001_013E': 'Income Bracket 12',
    'B19001_014E': 'Income Bracket 13',
    'B19001_015E': 'Income Bracket 14',
    'B19001_016E': 'Income Bracket 15',
    'B19001_017E': 'Income Bracket 16',
    'state': 'State',
    'county': 'County',
    'tract': 'Census Tract',
    'year': 'Year',
    'STATE_COUNTY_FIPS': 'State-County FIPS Code',
    'GEOID': 'Geographic Identifier',
    'FIPS': 'FIPS Code',
    'RPL_THEMES': 'SVI Themes',
    'CZ_FIPS': 'Combined Statistical Area FIPS',
    'ST': 'State Abbreviation',
    'LOCATION': 'Location',
    'E_TOTPOP': 'Estimated Total Population',
    'M_TOTPOP': 'Margin of Error Total Population',
    'E_HU': 'Estimated Housing Units',
    'M_HU': 'Margin of Error Housing Units',
    'E_UNEMP': 'Estimated Unemployed',
    'M_UNEMP': 'Margin of Error Unemployed',
    'E_LIMENG': 'Estimated Limited English Proficiency',
    'M_LIMENG': 'Margin of Error Limited English Proficiency',
    'E_MUNIT': 'Estimated Multi-Unit Housing',
    'M_MUNIT': 'Margin of Error Multi-Unit Housing',
    'E_MOBILE': 'Estimated Mobile Homes',
    'M_MOBILE': 'Margin of Error Mobile Homes',
    'E_CROWD': 'Estimated Crowded Housing',
    'M_CROWD': 'Margin of Error Crowded Housing',
    'E_NOVEH': 'Estimated No Vehicle',
    'M_NOVEH': 'Margin of Error No Vehicle',
    'DAMAGE_PROPERTY': 'Damage to Property',
    'DAMAGE_CROPS': 'Damage to Crops',
    'BEGIN_YEARMONTH': 'Begin Year-Month',
    'LAT': 'Latitude',
    'LON': 'Longitude',
    'storm_county': 'Storm County'
}

def load_hurricane_data():
    dfs = []
    for year in range(2010, 2023):
        path = f'../data/df_clean/df_hurricane_{year}.csv'
        try:
            df = pd.read_csv(path)
            df = df.rename(columns=census_svi_column_mapping)
            df['Year'] = year
            dfs.append(df)
        except FileNotFoundError:
            print(f" Skipping {year}: file not found")
    full = pd.concat(dfs, ignore_index=True)
    print(f" Loaded full_data with shape: {full.shape}")
    return full
```

```python
# Keep only the three features + target, drop any rows with NA
def prepare_features(df):
    features = [
        'Total Population',
        'Household Income Distribution',
        'Median Household Income'
    ]
    subset = df[features + ['Damage to Property']].dropna()
    print(f" After prepare_features: {subset.shape}")
    return subset
```

```python
# Pipeline
full_data = load_hurricane_data()
raw_train = full_data[full_data['Year'] < 2022].reset_index(drop=True)
raw_test  = full_data[full_data['Year'] == 2022].reset_index(drop=True)
train_df  = prepare_features(raw_train)
test_df   = prepare_features(raw_test)

X_train = train_df.drop('Damage to Property', axis=1)
y_train = np.log1p(train_df['Damage to Property'])
X_test  = test_df .drop('Damage to Property', axis=1)
y_test  = np.log1p(test_df ['Damage to Property'])

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

X_train_t = torch.tensor(X_train_scaled, dtype=torch.float32)
y_train_t = torch.tensor(y_train.values,     dtype=torch.float32)
X_test_t  = torch.tensor(X_test_scaled,  dtype=torch.float32)
y_test_t  = torch.tensor(y_test.values,     dtype=torch.float32)
```

```python
# Pyro model & guide with .to_event(1) on weights
def model(X, y=None):
    n = X.shape[1]
    intercept = pyro.sample("intercept", dist.Normal(0, 10))
    weights   = pyro.sample(
        "weights",
        dist.Normal(torch.zeros(n), 10 * torch.ones(n)).to_event(1)
    )
    sigma_raw = pyro.sample("sigma_raw", dist.HalfNormal(1.0))
    sigma     = 0.1 + sigma_raw
    mu = intercept + (X * weights).sum(dim=-1)
    with pyro.plate("data", X.shape[0]):
        pyro.sample("obs", dist.Normal(mu, sigma), obs=y)

def guide(X, y=None):
    n = X.shape[1]
    intercept_loc   = pyro.param("intercept_loc",    torch.tensor(0.0))
    intercept_scale = pyro.param("intercept_scale",  torch.tensor(1.0),
                                  constraint=dist.constraints.positive)
    weights_loc     = pyro.param("weights_loc",      torch.zeros(n))
    weights_scale   = pyro.param("weights_scale",    torch.ones(n),
                                  constraint=dist.constraints.positive)
    sigma_loc       = pyro.param("sigma_loc",        torch.tensor(1.0),
                                  constraint=dist.constraints.positive)

    pyro.sample("intercept",
                dist.Normal(intercept_loc, intercept_scale))
    pyro.sample(
        "weights",
        dist.Normal(weights_loc, weights_scale).to_event(1)
    )
    pyro.sample("sigma_raw",
                dist.HalfNormal(sigma_loc))
```

```python
# Clear old params & train with loss tracking
pyro.clear_param_store()
optimizer = Adam({"lr": 0.01})
svi       = SVI(model, guide, optimizer, loss=Trace_ELBO())

num_epochs = 2000
train_losses = []
test_losses = []

for epoch in range(1, num_epochs + 1):
    # Training step
    train_loss = svi.step(X_train_t, y_train_t)
    train_losses.append(train_loss)
    # Evaluate on test set
    test_loss = svi.evaluate_loss(X_test_t, y_test_t)
    test_losses.append(test_loss)
    if epoch % 500 == 0:
        print(f" Epoch {epoch:4d} | Train Loss: {train_loss:.2f} | Test Loss: {test_loss:.2f}")
```

```python
# Plot loss curves
plt.figure(figsize=(8, 5))
epochs = list(range(1, num_epochs + 1))
plt.plot(epochs, train_losses, label='Train Loss')
plt.plot(epochs, test_losses,  label='Test Loss')
plt.xlabel('Epoch')
plt.ylabel('ELBO Loss')
plt.title('Training vs. Test Loss Over Epochs')
plt.legend()
plt.show()
```

```python
# Posterior coefficients barchart
# get learned variational parameters
w_loc   = pyro.param("weights_loc").detach().numpy()
w_scale = pyro.param("weights_scale").detach().numpy()
features = X_train.columns.tolist()

plt.figure(figsize=(6,4))
plt.bar(features, w_loc, yerr=w_scale, capsize=5)
plt.xticks(rotation=30, ha="right")
plt.ylabel("Weight mean  1")
plt.title("Variational posterior for each feature weight")
plt.tight_layout()
plt.show()
```

```python
# Predicted vs. Actual scatter 
# Note: this doesn't make much sense here since we are predicitng only 1 year
# reconstruct point predictions
b0 = pyro.param("intercept_loc").item()
w  = pyro.param("weights_loc").detach().numpy()
y_pred = np.expm1(b0 + X_test_t.numpy() @ w)
y_true = np.expm1(y_test_t.numpy())

plt.figure(figsize=(5,5))
plt.scatter(y_true, y_pred, alpha=0.3, s=5)
lims = [y_true.min(), y_true.max()]
plt.plot(lims, lims, 'k--')  # 45 line
plt.xlabel("Actual Damage")
plt.ylabel("Predicted Damage")
plt.title("Predicted vs. Actual (2022)")
plt.tight_layout()
plt.show()
```

```python
# Residual vs each feature
# after training set b0, w_loc, and built X_test_t/y_test_t

#Random scatter around0: model fits well.  
#Up/down trend: missing nonlinearity; add transforms.  
#Fan shape: heteroscedasticity; consider variance modeling.  
#Clusters/outliers: data issues or omitted variables.
    
# Reconstruct point predictions on the test set:
b0    = pyro.param("intercept_loc").item()
w_loc = pyro.param("weights_loc").detach().numpy()

# y_pred is on the original scale
y_pred_log = b0 + X_test_t.numpy() @ w_loc
y_pred     = np.expm1(y_pred_log)

# y_true likewise back-transform
y_true     = np.expm1(y_test_t.numpy())

# Compute residuals
residuals = y_true - y_pred

for i, feat in enumerate(features):
    plt.figure(figsize=(5,4))
    plt.scatter(X_test.iloc[:,i], residuals, alpha=0.3, s=5)
    plt.axhline(0, color='k', linestyle='--')
    plt.xlabel(feat)
    plt.ylabel("Residual")
    plt.title(f"Residual vs. {feat}")
    plt.tight_layout()
    plt.show()
```

```python
# assume X_test, scaler, b0, and w are already defined in the environment
features = X_test.columns.tolist()
median_vals = X_test.median()

for feat in features:
    # create grid for this feature
    grid = np.linspace(X_test[feat].min(), X_test[feat].max(), 100)
    
    # build a DataFrame where every column is at its median except 'feat'
    X_pd = pd.DataFrame(
        np.tile(median_vals.values, (100, 1)),
        columns=features
    )
    X_pd[feat] = grid
    
    # scale and convert to tensor
    X_pd_scaled = scaler.transform(X_pd)
    X_pd_t = torch.tensor(X_pd_scaled, dtype=torch.float32)
    
    # predict and back-transform
    y_pd = np.expm1(b0 + X_pd_t.numpy() @ w)
    
    # plot partial dependence
    plt.figure(figsize=(5, 4))
    plt.plot(grid, y_pd)
    plt.xlabel(feat)
    plt.ylabel("Predicted Damage")
    plt.title(f"Partial Dependence on {feat}")
    plt.tight_layout()
    plt.show()
```

```python
# Weights posterior samples (uncertainty)
# sample 200 draws of weights
w_scale = pyro.param("weights_scale").detach().numpy()
w_samples = np.random.normal(w_loc, w_scale, size=(200, len(w_loc)))

plt.figure(figsize=(6,4))
for i, feat in enumerate(features):
    plt.violinplot(w_samples[:,i], positions=[i], widths=0.6)
plt.xticks(range(len(features)), features, rotation=30, ha="right")
plt.ylabel("Weight value")
plt.title("Approximate posterior via variational samples")
plt.tight_layout()
plt.show()
```
